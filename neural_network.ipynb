```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import pandas as pd

from google.colab import drive

drive.mount('/content/drive')
```

```python
michigan_df = pd.read_csv('michigan_df.csv')
```

```python
michigan_df.head()
michigan_df = michigan_df.drop(columns='Unnamed: 0')
```

```python
X_encoded = pd.get_dummies(michigan_df,dtype=int)
```

```python
X_encoded.head()
```

```python
y = michigan_df['price']
X = michigan_df.drop(columns='price')

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
X_train.dtypes
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
print(X_train_scaled.shape)
```

```python
nn = tf.keras.models.Sequential()
```

```python
nn.add(tf.keras.layers.Dense(units=64,activation='relu',input_shape=(X_train.shape[1],)))
```

```python
nn.add(tf.keras.layers.Dense(units=32,activation='relu'))
```

```python
nn.add(tf.keras.layers.Dense(units=1))
```

```python
nn.summary()
```

```python
nn.compile(loss='mean_squared_error',optimizer='adam',metrics=['accuracy'])
```

```python
fit_model = nn.fit(X_train_scaled,y_train,epochs=500)
```

```python
y_pred = nn.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn import metrics
```

```python
mse = mean_squared_error(y_test, y_pred)

# Calculate Root Mean Squared Error
rmse = np.sqrt(mse)
accuracy = nn.evaluate(X_test_scaled, y_test)
print(accuracy)
print(f"Root Mean Squared Error (RMSE): {rmse}")
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
```
