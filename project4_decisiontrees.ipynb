```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from google.colab import drive
drive.mount('/content/drive')
```

```python
csv_file_path = "/content/drive/My Drive/realtor-data.zip.csv"
```

```python
df = pd.read_csv(csv_file_path)
```

```python
clean_df = df.dropna()
clean_df = clean_df.drop_duplicates()
clean_df.head()
```

```python
michigan_df = clean_df[clean_df['state']=='Michigan']
michigan_df.head()
michigan_df.shape
```

```python
michigan_df = michigan_df.drop(columns=['state'])
michigan_df.head()
michigan_df = michigan_df.drop(columns=['status'])
```

```python
michigan_df.head()
```

```python
michigan_df = michigan_df.drop(columns=['street'])
```

```python
michigan_df = michigan_df.drop(columns=['city'])
```

```python
# print(df.dtypes)
```

```python
label_encoders = {}
for column in ['brokered_by','zip_code','city']:
  label_encoders[column]=LabelEncoder()
  michigan_df[column]=label_encoders[column].fit_transform(michigan_df[column])
```

```python
feature_columns = ['bed','bath','acre_lot','house_size','zip_code']
target_column = 'price'
X = michigan_df[feature_columns]
y = michigan_df[target_column]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2,random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train_scaled,y_train)
```

```python
predictions = dt_model.predict(X_test_scaled)
```

```python
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print(f"MSE: {mse}")
print(f"r2: {r2}")
```
